### 5.1 Выберите дом клиенту

Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

C видом на море, подвалом и тремя или больше спальнями получается 41 вариант, отсортированный по цене:

```python
homeland[(homeland['bathrooms'] >= 3) & (homeland['sqft_basement'] > 0) & (homeland['waterfront'] > 0)].sort_values(['price'])
```

### 5.2 Выберите дом клиенту
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?


    посмотрим на сколько хорошие бывают виды

```python
homeland['view'].value_counts().sort_values(ascending=False)
```

Подсчитаем подходящие варианты с датой постройки 1980 или моложе, в очень хорошем состоянии и видом на набережную или очень хорошим видом -- их получается 26.

```python
var5_2 = homeland[(homeland['yr_built'] >= 1980) & (homeland['condition'] >= 4) & ((homeland['waterfront'] > 0) | (homeland['view'] == 4))].sort_values(['price'])
var5_2.shape
```

```python
print(f" Диапазон цен на подходящие варианты составляет от $ {var5_2['price'].min():,} до $ {var5_2['price'].max():,} ")
```
