### Задача 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

### Задача 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition')['price'].agg(['min','mean','max'])
```

### Задача 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
pd.set_option('display.float_format', '{:,}'.format)
df.pivot_table(index='waterfront', columns='view', values='id', aggfunc='count', fill_value=0)
```

### Задача 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pv = df.pivot_table(index='floors', columns='bedrooms', values='id', aggfunc='count', fill_value=0)
pv_max = pv.max().max()
floors_w_max = pv[pv == pv_max].stack().index[0]
print(pv)
print(f"Максимальное значение домов: {pv_max} при количестве спален - {floors_w_max[1]} и этажности - {floors_w_max[0]:.0f}")
```

### Задача 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.pivot_table(index='condition', columns='grade', values='price', aggfunc='median', fill_value=0)
```

### Задача своя. Стало интересно посчитать при каком сочетании этажности и количестве спальных комнат в доме максимальная средняя цена 

с учетом того, что уникальные предложения (малое их количество) могут испортить статистику, берем количество интересующих сочетаний 10 или больше

```python
df_filter = df.groupby(['floors', 'bedrooms']).filter(lambda x: len(x) >= 10)
pv = df_filter.pivot_table(index='floors', columns='bedrooms', values='price', aggfunc='median', fill_value=0)
pv_max = pv.max().max()
floors_w_max = pv[pv == pv_max].stack().index[0]
print(pv)
print(f"Макисмальное среднее значние цены: {pv_max:,} при количестве спален - {floors_w_max[1]:} и этажности - {floors_w_max[0]}")
```
